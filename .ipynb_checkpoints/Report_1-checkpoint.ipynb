```python
from sklearn import cluster
from sklearn.preprocessing import LabelEncoder
n_samples, n_features = scaled_data.shape

sillhoute = []
completenes = []
homogeneity = []

wcss = []

n_digits = len(np.unique(Y))
Y2 = LabelEncoder().fit_transform(Y)
for k in range(2, 30):
 kmeans = cluster.KMeans(n_clusters=k)
 y_fit = kmeans.fit_predict(scaled_data)
 print(k)
#  print(y_fit)
#  print(metrics.silhouette_score(scaled_data, kmeans.labels_))
#  print(metrics.completeness_score(Y2, kmeans.labels_))
#  print(metrics.homogeneity_score(Y2, kmeans.labels_))
#  print("   ")
 sillhoute.append(metrics.silhouette_score(scaled_data, kmeans.labels_))
 completenes.append(metrics.completeness_score(Y2, kmeans.labels_))
 homogeneity.append(metrics.homogeneity_score(Y2, kmeans.labels_))
 wcss.append(kmeans.inertia_)
```

```python
plt.plot(sillhoute,'g*', completenes, 'ro', homogeneity, '--')
plt.show()
```

```python
plt.plot(wcss, marker='x')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = cluster.KMeans(n_clusters=5).fit(scaled_data)
# y_fit = kmeans.fit(scaled_data)
dataset['labels'] = kmeans.labels_
# print(kmeans.labels_)
# kmeans.cluster_centers_
print("Silhouette score", metrics.silhouette_score(scaled_data, kmeans.labels_))
print("Homogeneity score", metrics.homogeneity_score(Y2, kmeans.labels_))
print("Completeness score", metrics.completeness_score(Y2, kmeans.labels_))
```

```python
##Single value linear regression
class12_df = dataset.drop(['Gender','10percentage','CollegeTier','Degree','Specialization','collegeGPA','English','Logical','Quant','Domain','ComputerProgramming','ElectronicsAndSemicon','ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg','conscientiousness','agreeableness','extraversion','nueroticism','openess_to_experience','Salary', 'labels'],axis='columns')
class10_df = dataset.drop(['Gender','CollegeTier', '12percentage','Degree','Specialization','collegeGPA','English','Logical','Quant','Domain','ComputerProgramming','ElectronicsAndSemicon','ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg','conscientiousness','agreeableness','extraversion','nueroticism','openess_to_experience','Salary', 'labels'],axis='columns')
salary = dataset.Salary
```

```python
reg = linear_model.LinearRegression()
reg.fit(class12_df,salary)
```

```python
# Annual salary actual and prediction comparison for 5 random students with class 12 percentage from dataset.
perecentage = []
actual_salary = []
predicted_salary = []

ranadom_df = dataset.sample(n=5)
for i, row in ranadom_df.iterrows():
    perecentage.append(row['12percentage'])
    actual_salary.append(row['Salary'])
    predicted_salary.append((reg.predict([[row['12percentage']]])).astype(int)[0])


X = np.arange(5)
width = 0.2
Gender=['Actual Salary','Predicted Salary']
fig, ax = plt.subplots()
ax.bar(X, actual_salary, width, color='blue')
ax.bar(X+width, predicted_salary, width, color='y')
ax.set_xticks(X + width / 2)
ax.set_xticklabels(perecentage)
plt.xlabel("12th Percentage")
plt.ylabel('Salary')
plt.title('Actual and Predected Salary')
plt.legend(Gender,loc=2)
plt.show()
```
